```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg', 'pdf')
sns.set()
%matplotlib inline
```

```python
entropy = lambda p: -p * np.log2(p) - (1 - p) * np.log2(1 - p)
cross_entropy_wrt_uniform_dist = lambda p: -0.5 * np.log2(p) - 0.5 * np.log2(1 - p) 

x_axis = np.arange(0.001, 1, 0.001)

neg_entropys = []
cross_entropys= []
for p in x_axis:
    neg_entropys.append(-entropy(p))
    cross_entropys.append(cross_entropy_wrt_uniform_dist(p))


plt.plot(x_axis, neg_entropys, label='negetive entropy of bernoulli distribution')
plt.plot(x_axis, cross_entropys, label='cross entropy wrt uniform distribution')
plt.xlabel('probability')
plt.legend()
plt.show()
```
